```python
from torch import nn
from torch.utils.data import DataLoader
from torchvision import transforms, datasets
import numpy as np
import pandas as pd
```

```python
class NeuralNetwork:
    def __init__(self, *args):
        self.layers = args
    
    def __call__(self, x):
        print(f"X : {x.shape}")
        for i in range(len(self.layers)):
            x = self.layers[i](x)
        return np.argmax(x)
    
    class Layer:
        def __init__(self, in_features = None, out_features = None, bias = True, activation = "None"):
            if activation == "None" and (in_features == None or out_features == None):
                raise KeyError("Features not set for Layer!")


            activation_functions = {"None" : None, "Sigmoid" : (lambda x : 1 / (1 + np.exp(-x))), "Softmax" : (lambda x, axis = 0 : np.exp(x) / np.sum(np.exp(x), axis = 0)), "ReLU" : (lambda x : np.where(x > 0, x, 0)), "Leaky ReLU" : lambda x, alpha = 0.05 : np.where(x > 0, x, x * alpha)}
            

            self.weights = None
            self.bias = None
            self.__dict = {activation}

            self.activation = activation_functions[activation]

            if activation == "None":
                self.weights = np.random.rand(in_features,out_features)
                self.bias = np.zeros(shape = out_features)
                if bias:
                    self.bias = np.random.rand(1,out_features)

                self.__dict = {"Weight" : self.weights,
                            "Bias" : self.bias}

        def __call__(self, *args):
            if self.activation != None:
                print((self.activation)(*args))                
                return (self.activation)(*args)
            else:
                print(np.add(np.matmul(*args,self.weights), self.bias))
                return np.add(np.matmul(*args,self.weights), self.bias)

                    
        
        def state_dict(self):
            return self.__dict
        
    def state_dict(self):
        self.__dict = {}
        for i in range(len(self.layers)):
            self.__dict[f"Layer {i}"] = self.layers[i].state_dict()
        
        return self.__dict
    
    def state_shape(self):
        self.__dict = {}
        for i in range(len(self.layers)):
            try:
                self.__dict[f"Layer {i}"] = {"Weight" : self.layers[i].state_dict()["Weight"].shape, "Bias" : self.layers[i].state_dict()["Bias"].shape}
            except:
                self.__dict[f"Layer {i}"] = self.layers[i].state_dict()
        
        return self.__dict
    
    def summary(self):
        self.__dict = {"Start" : f"1 x {list(self.state_shape()["Layer 0"]["Weight"])[0]}",
                       "End" : f"1 x {list(self.state_shape()[f"Layer {len(self.layers) - 2}"]["Weight"])[1]}"
                       }
        
        return self.__dict
```

```python
ai = NeuralNetwork(
    NeuralNetwork.Layer(3,6), 
    NeuralNetwork.Layer(activation= "ReLU"), 
    NeuralNetwork.Layer(6,8),
    NeuralNetwork.Layer(activation= "ReLU"), 
    NeuralNetwork.Layer(8,4),
    NeuralNetwork.Layer(activation= "ReLU"), 
    NeuralNetwork.Layer(4,3),
    NeuralNetwork.Layer(activation= "Softmax"), 
    )
```
